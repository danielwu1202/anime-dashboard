```python
import pandas as pd
import numpy as np
import collections
```

```python
origin_anime_data = pd.read_csv('Anime.csv')
origin_anime_data.head()
```

```python
# 查看欄位並刪除不要的欄位
anime_data = origin_anime_data.drop(['Synonyms', 'Premiered', 'Broadcast'], axis=1)
anime_data.columns
```

```python
# 按照日期排序
anime_data = anime_data.sort_values(by='Start_Aired')
anime_data.head()
```

```python
# 將年份擷取出來形成新columns
anime_data['year'] = anime_data['Start_Aired'].str.extract(r'(\d{4})')
anime_data['year'] = pd.to_numeric(anime_data['year'])
```

```python
counter_year = collections.Counter(anime_data["year"])
df_year = pd.DataFrame({"year":counter_year.keys(),
                    "count": counter_year.values()})
```

```python
df_year[df_year["count"] > 100]
```

```python
# 統計基本資料
counter_demographics = collections.Counter(anime_data["Demographics"])
print("觀看族群", counter_demographics)
```

```python
# 將日文轉換成英文
anime_data.replace(['Shounen', 'Seinen', 'Shoujo', 'Josei'], ['young boy', 'teenager', 'young girl', 'female'], inplace=True)
```
