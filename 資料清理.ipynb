```python
origin_anime_data = pd.read_csv('Anime.csv')
origin_anime_data.head()
```

```python
# 查看欄位並刪除不要的欄位
anime_data = origin_anime_data.drop(['Synonyms', 'Premiered', 'Broadcast'], axis=1)
anime_data.columns
```

```python
# 按照日期排序
anime_data = anime_data.sort_values(by='Start_Aired')
anime_data.head()
```

```python
# 將年份擷取出來形成新columns
anime_data['year'] = anime_data['Start_Aired'].str.extract(r'(\d{4})')
anime_data['year'] = pd.to_numeric(anime_data['year'])
```

```python
counter_year = collections.Counter(anime_data["year"])
df_year = pd.DataFrame({"year":counter_year.keys(),
                    "count": counter_year.values()})
```

```python
df_year[df_year["year"] > 1990]
```

```python
df_year[df_year["count"] > 50]
```

```python
# 統計基本資料
counter_demographics = collections.Counter(anime_data["Demographics"])
print("觀看族群", counter_demographics)
```

```python
# 將日文轉換成英文
anime_data.replace(['Shounen', 'Seinen', 'Shoujo', 'Josei'], ['young boy', 'teenager', 'young girl', 'female'], inplace=True)
```

```python
counter_type = collections.Counter(anime_data['Type'])
print(counter_type)
```

### 依據type分析

```python
TV_mask = anime_data["Type"] == "TV"
OVA_mask = anime_data['Type'] == 'OVA'
movie_mask = anime_data['Type'] == 'Movie'
ONA_mask = anime_data['Type'] == 'ONA'
special_mask = anime_data['Type'] == 'Special'
music_mask = anime_data['Type'] == 'Music'
unknown_type_mask = anime_data['Type'] == 'Unknown'
```

```python
# TV
TV = anime_data[TV_mask]
TV.to_csv('TV_anime.csv', index=False)
```

```python
# OVA
OVA = anime_data[OVA_mask]
OVA.to_csv('OVA_anime.csv', index=False)
```

```python
# movie
movie = anime_data[movie_mask]
movie.to_csv('movie_anime.csv', index=False)
```

```python
# ONA
ONA = anime_data[ONA_mask]
ONA.to_csv('ONA_movie.csv', index=False)
```

```python
# special
special = anime_data[special_mask]
special.to_csv('special_anime.csv', index=False)
```

```python
# music
music = anime_data[music_mask]
music.to_csv('music_anime.csv', index=False)
```

```python
# unknown
unknown = anime_data[unknown_type_mask]
unknown.to_csv('unknown_type_anime.csv', index=False)
```
