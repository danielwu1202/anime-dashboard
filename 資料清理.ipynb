```python
origin_anime_data = pd.read_csv('Anime.csv')
origin_anime_data.head()
```

```python
origin_anime_data.rename(columns={'ID':'anime_ID'}, inplace=True)
```

```python
# 查看欄位並刪除不要的欄位
anime_data = origin_anime_data.drop(['Synonyms', 'Premiered', 'Broadcast'], axis=1)
anime_data.columns
```

```python
anime_data['Start_Aired'] = pd.to_datetime(anime_data['Start_Aired'], errors='coerce')
anime_data['End_Aired'] = pd.to_datetime(anime_data['End_Aired'], errors='coerce')
```

```python
anime_data.sort_values(by='Start_Aired', inplace=True)
```

```python
anime_data
```

```python
# 將日文轉換成英文
anime_data.replace(['Shounen', 'Seinen', 'Shoujo', 'Josei', 'Kids, Shounen', 'Kids, Shoujo', 'Josei, Shoujo', 'Kids, Seinen'], ['young boy', 'teenager', 'young girl', 'female', 'kids, young boy', 'kids, young girl', 'female, young girl', 'kids, teenager'], inplace=True)
```

```python
# 移除2023年以後的資料
anime_data = anime_data[anime_data['Start_Aired'] < '2023']
```

```python
anime_data = anime_data.drop(anime_data[anime_data['Type'] == 'Unknown'].index)
```

```python
anime_data.to_csv('anime_data.csv', index=False, encoding='utf-8')
```

```python
# 統計基本資料
counter_demographics = collections.Counter(anime_data["Demographics"])
print("觀看族群", counter_demographics)
```
